```python
# Data Cleaning and Preprocessing
# Checking Nan values
df.isnull().sum()
```

```python
# checking for special(-1) values
(df == -1).sum()
```

```python

# replace -1 with NaN
df.replace(-1, np.nan, inplace=True)
```

```python
# Replace NaN values with the mean of the respective column
df=df.fillna(df.mean(numeric_only=True))
```

```python
# Check for missing values again
df.isnull().sum()
```

```python
# Cleaning the column names
df.columns = (df.columns
.str.strip()
.str.lower()
.str.replace(' ', '_')
.str.replace(r'\(.*?\)', '', regex=True)  # remove anything in parentheses
.str.replace(r'1000_ha|1000_tons|kg_per_ha', '', regex=True)  # remove specific units
.str.replace(r'[^a-z0-9_]', '', regex=True)  # keep alphanumeric + underscore
.str.replace(r'__+', '_', regex=True)  # collapse double underscores
.str.strip('_')  # remove leading/trailing underscores
)
```

```python
# Display the cleaned column names
df.columns
```

```python
# Removing column names with NaN
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
```

```python
# Check for null values and duplicates
df.isnull().sum(), df.duplicated().sum()
```

```python
# Download cleaned dataset
df.to_csv(r'D:\Projects\Agriproj\ICRISAT-District Level Data - Cleaned.csv', index=False)
```

```python
# 2: Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

# Step 1: Group and get Top 5 Wheat Producing States
top_wheat_states = (
    df.groupby('state_name')['wheat_production'] 
    .sum() # Grouping by state and summing wheat production
    .sort_values(ascending=False) # Sort by wheat production in descending order
    .head(5) 
)


# Step 2: Prepare DataFrame for plotting
wheat_df = top_wheat_states.reset_index()
wheat_df.columns = ['state', 'wheat_production']

# Step 3: Calculate percentage for pie chart
wheat_df['percentage'] = (wheat_df['wheat_production'] / wheat_df['wheat_production'].sum()) * 100

# ---------- PLOT 1: Bar Chart ----------
plt.figure(figsize=(10, 5))
sns.barplot(data=wheat_df, x='state', y='wheat_production', palette='inferno')  # Using a color palette for better aesthetics
plt.title('Top 5 Wheat Producing States in India', fontsize=14)
plt.xlabel('State')
plt.ylabel('Wheat Production (1000 tons)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------- PLOT 2: Pie Chart ----------
plt.figure(figsize=(8, 6))
plt.pie(wheat_df['wheat_production'], labels=wheat_df['state'], autopct='%1.1f%%', colors=sns.color_palette('Oranges'), startangle=140)
plt.title('Percentage Share of Top 5 Wheat Producing States')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
# 4: Top 7 SUNFLOWER PRODUCTION  State

top_sunflower_states = (
    df.groupby('state_name')['sunflower_production'] 
    .sum() # Grouping by state and summing sunflower production
    .sort_values(ascending=False) # Sort by sunflower production in descending order
    .head(7) 
)

# Step 2: Prepare DataFrame for plotting
top_sunflower_df = top_sunflower_states.reset_index()
top_sunflower_df.columns = ['state', 'sunflower_production']

# Step 3: Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.barplot(
    data=top_sunflower_df,
    x='state',
    y='sunflower_production',
    palette='YlOrBr',  # Yellow-Orange-Brown: fits sunflower theme 🌻
    edgecolor='black'
)

# Step 4: Add value labels on bars
for index, row in top_sunflower_df.iterrows():
    plt.text(index, row.sunflower_production + 10, f"{row.sunflower_production:.0f}",
             ha='center', va='bottom', fontsize=10)

# Step 5: Formatting
plt.title('Top 7 Sunflower Producing States in India', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Sunflower Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
#5: India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

# Step 1: Filter data for the last 50 years
last_50_years = df[df['year'] >= (df['year'].max() - 50)]
# Step 2: Group by year and sum sugarcane production
sugarcane_yearly = last_50_years.groupby('year')['sugarcane_production'].sum().reset_index()
# Step 3: Plot using Seaborn + Matplotlib
plt.figure(figsize=(12, 6))
sns.lineplot(data=sugarcane_yearly, x='year', y='sugarcane_production', marker='o', linestyle='-', linewidth=2.5, color='brown')
plt.title('India\'s Sugarcane Production Over the Last 50 Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Sugarcane Production (1000 tons)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#6: Rice Production Vs Wheat Production (Last 50y)

# Step 1: Filter last 50 years
last_50_years = df[df['year'] >= (df['year'].max() - 50)]

# Step 2: Group by year and sum rice and wheat production
production_trend = (
    last_50_years.groupby('year')[['rice_production', 'wheat_production']]
    .sum()
    .reset_index()
    .sort_values('year')
)

# Step 3: Plot
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Rice line
sns.lineplot(
    data=production_trend,
    x='year',
    y='rice_production',
    label='Rice Production',
    marker='o',
    color='seagreen'
)

# Wheat line
sns.lineplot(
    data=production_trend,
    x='year',
    y='wheat_production',
    label='Wheat Production',
    marker='s',
    color='peru'
)

# Step 4: Formatting
plt.title('Rice vs Wheat Production in India (Last 50 Years)', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Crop')
plt.tight_layout()
plt.show()

```

```python
#7: Rice Production By West Bengal Districts


# Step 1: Filter data for West Bengal
wb_rice = df[df['state_name'].str.lower() == 'west bengal']

# Step 2: Group by district and sum rice production
wb_districts = (
    wb_rice.groupby('dist_name')['rice_production']
    .sum()
    .reset_index()
    .sort_values('rice_production', ascending=False)
)

# Step 3: Plot
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(
    data=wb_districts,
    x='dist_name',
    y='rice_production',
    palette='Greens_r',
    edgecolor='black'
)

# Step 4: Formatting
plt.title('Total Rice Production by District in West Bengal', fontsize=14)
plt.xlabel('District', fontsize=12)
plt.ylabel('Rice Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#8: Top 10 Wheat Production Years From UP

# Step 1: Filter data for Uttar Pradesh
up_wheat = df[df['state_name'].str.lower() == 'uttar pradesh']

# Step 2: Group by year and sum wheat production
up_wheat_yearly = (
    up_wheat.groupby('year')['wheat_production']
    .sum()
    .reset_index()
    .sort_values('wheat_production', ascending=False)
    .head(10)
    .sort_values('year')  # Optional: sort chronologically for better visuals
)


# Step 3: Plot
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.barplot(
    data=up_wheat_yearly,
    x='year',
    y='wheat_production',
    palette='Blues_d',
    edgecolor='black'
)

# Step 4: Formatting
plt.title('Top 10 Wheat Production Years in Uttar Pradesh', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Wheat Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#9: Millet Production (Last 50y)    

# Step 1: Filter last 50 years
last_50_years = df[df['year'] >= (df['year'].max() - 50)]

# Step 2: Group and sum by year
millet_trend = (
    last_50_years.groupby('year')[['pearl_millet_production', 'finger_millet_production']]
    .sum()
    .reset_index()
    .sort_values('year')
)

# Step 3: Plot using matplotlib
plt.figure(figsize=(12, 6))
plt.plot(millet_trend['year'], millet_trend['pearl_millet_production'],
         label='Pearl Millet', color='goldenrod', marker='o', linewidth=2)

plt.plot(millet_trend['year'], millet_trend['finger_millet_production'],
         label='Finger Millet', color='purple', marker='s', linewidth=2)

# Step 4: Labels and formatting
plt.title("Pearl vs Finger Millet Production in India (Last 50 Years)", fontsize=15)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Production (1000 tons)", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
